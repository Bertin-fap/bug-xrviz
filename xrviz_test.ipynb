## issue with xrviz
using the following script we obtain:

|unit_x|unit_y|unit_lbd|plot working?|
|-|-|-|-|
|u.a.|u.a.|u.a.|Y|
|u.a.|u.a.|cm-1|N|
|µm|µm|u.a.|Y|
|cm-1|µm|u.a.|N|

so the unit cm-1 which is quit usual in the world of spectroscopy is not correctly interpreted.

```python
import xarray as xr
import numpy as np
from xrviz.dashboard import Dashboard
import matplotlib.pyplot as plt

Intensity = np.random.rand(15,15,50)
unit_x = 'cm-1'
unit_y = 'µm'
unit_lbd = 'u.a.'
x = np.arange(15, dtype = float)
y = np.arange(15, dtype = float)
lbd = np.arange(50, dtype = float)

da = xr.DataArray(
                Intensity,
                dims=["y", "x","lbd"],
                name='Intensity',
                attrs={'units':'u.a.'},
                coords={
                            "y": xr.DataArray(x,name = 'y',
                                 dims = ['y'],
                                 attrs = {'units': unit_y}),
                            "x": xr.DataArray(y,name = 'x',
                                 dims = ['x'],
                                 attrs={'units': unit_x}),
                            "lbd" : xr.DataArray(lbd, name = 'lbd',
                                 dims = ['lbd'],
                                 attrs = {'units': unit_lbd})
                       }
             )

dash = Dashboard(da )
dash.panel
```
